# Correlación y covarianza

Es importante identificar variables que estén correlacionadas para poder evitar multicolinealidad en un marco de datos.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species')
plt.show()
```

```python
iris.columns
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled
```

Para poder calcular la matriz de covarianza entre unos datos escalados, en necesario obtener la transpuerta.

```python
sc_t = scaled.T
```

```python
cov_matr = np.cov(sc_t)
cov_matr
```

```python
sns.heatmap(cov_matr,
            cbar = True,
            annot = True,
            square = True,
            fmt = '.2f',
            annot_kws = {'size': 12},
            yticklabels = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
            xticklabels = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.show()
```
